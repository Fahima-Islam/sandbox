```python
# some goodies
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import math
from mpl_toolkits.mplot3d import Axes3D
```

```python
# the initial energy, Ei= 100 meV
# the range of scattering angle (phi) is from 0 to 160 degree
```

```python
# Relationship between Q, E and phi
# Q= 0.6947 sqrt (2Ei-E-2 sqrt(Ei(Ei-E))cos phi)
```

```python
fig = plt.figure()
ax = Axes3D(fig)
Energy=np.arange(-100,100)
Angle=np.arange(0, 160)
E, phi=np.meshgrid(Energy, Angle)
Q=0.6947*np.sqrt ((2*100)-E-(2*np.sqrt(100*(100-E)))*np.cos(phi))


ax.plot_surface(E,phi,Q,rstride=1, cstride=1, cmap='hot')

ax.set_xlabel('E')
ax.set_ylabel('phi')
ax.set_zlabel('Q')

plt.show()
```
