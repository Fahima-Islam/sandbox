```python
# some goodies
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
# the initial energy, Ei
# the range of scattering angle (phi) is from 0 to 140 degree with 20 degree increament
```

```python
# Relationship between Q, E and phi
# Q= 0.6947 sqrt (2Ei-E-2 sqrt(Ei(Ei-E))cos phi)
```

```python
Ei=input('Enter the initial value of energy: ')
```

```python
Energy=np.arange(-Ei,Ei)
Angle=np.arange(0,140,20)
    
for phi in Angle:
    result=[]
    for E in Energy:
        Q=57.2958*(0.6947*math.sqrt ((2*Ei)-E-(2*math.sqrt(Ei*(Ei-E)))*math.cos(phi)))
        result.append(Q)  
    plt.plot(result,Energy,label='$phi={phi}deg$'.format(phi=phi))
plt.xlabel('Q [A^-1]')
plt.ylabel('E [meV]')
plt.legend(loc='best')
plt.show()
```
