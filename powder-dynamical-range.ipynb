```python
# some goodies
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
# the initial energy, Ei
# the range of scattering angle (phi) is from 0 to 140 degree with 20 degree increament
```

```python
# Relationship between Q, E and phi
# Q= 0.6947 sqrt (2Ei-E-2 sqrt(Ei(Ei-E))cos phi)
```

```python
Ei=input('Enter the initial value of energy: ')
```

```python
## Calculation of dynamical range with numpy library
Energy=np.arange(-Ei,Ei)
Angle=np.arange(0,140,20)
E,phi=np.meshgrid(Energy,Angle)
Q=0.6947*np.sqrt ((2*Ei)-E-(2*np.sqrt(Ei*(Ei-E)))*np.cos(np.deg2rad(phi)))

for i in xrange(len(Angle)):
    ang=Angle[i]
    plt.plot(Q[i][:],Energy,label='$phi={ang}deg$'.format(ang=ang) )
plt.xlabel('Q [A^-1]')
plt.ylabel('E [meV]')
plt.legend(loc='best')
plt.show()


```

```python
## Calculation of dynamical range without numpy library
for phi in Angle:
    result=[]
    for E in Energy:
        Q=0.6947*math.sqrt ((2*Ei)-E-(2*math.sqrt(Ei*(Ei-E)))*math.cos(math.radians(phi)))
        result.append(Q)  
    plt.plot(result,Energy,label='$phi={phi}deg$'.format(phi=phi))
plt.xlabel('Q [A^-1]')
plt.ylabel('E [meV]')
plt.legend(loc='best')
plt.show()
```
